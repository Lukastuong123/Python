````python
#What is the average tip amount
a=tips.tip.mean()
print('Average tip amount is: ',round(a,3))
print('---------')

#What is the average bill amount
b=tips.total_bill.mean()
print('Average total_bill is: ',round(b,3))
print('---------')

#What is the average tip by gender
c=tips.groupby(['sex'])
c1=c.get_group('Male')['tip'].mean()
c2=c.get_group('Female')['tip'].mean()
print('Average tip among male is: ',round(c1,3))
print('Average tip among female is: ',round(c2,3))
print('---------')

#What is the average tip on a sunday by gender
Df = tips.groupby(['sex', 'day'])
d1=Df.get_group(('Male', 'Sun'))['tip'].mean()
d2=Df.get_group(('Female', 'Sun'))['tip'].mean()
print('Average tip among male is: ',round(d1,3))
print('Average tip among female is: ',round(d2,3))
print('---------')




#Which party size gives the maximum tip as a percent of total bill amount
Ef = tips.groupby(['size'])
list(Ef)[0]
# max_value=0
# for i in range(int(min(Ef['size'])),int(max(Ef['size']))+1):
#     e1= Ef.get_group(i)['tip'].sum()
#     e2= Ef.get_group(i)['total_bill'].sum()
#     a= e1/e2*100
#     if a >max_value: 
#         max_value==a
#     print('The size gives the maximum tip as a percent of total bill amount: ',i)
````
